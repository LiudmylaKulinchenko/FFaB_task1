## Побудова Графіка Лінії Тренду

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

1) Завантаження даних з умови задачі

```python
years = np.array([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021])
revenue = np.array([
        223624700.0,
        233624700.0, 
        387652300.0,
        443012100.0, 
        593168100.0, 
        668893400.0, 
        869642100.0, 
        673978800.0,
        797978900.0
    ])
```

2) Побудова та відображення графіку лінії треду

```python
plt.figure(figsize=(10, 5))
plt.plot(years, revenue, marker='o', linestyle='-', color='b', label='Виручка')
plt.xlabel('Роки')
plt.ylabel('Виручка, грн.')
plt.title('Фінансовий прогноз')
plt.grid(True)
plt.legend()
```

### Пояснення результатів:

- З графіка видно, що виручка зростала протягом років. Це може вказувати на *позитивну динаміку бізнесу або ринку*.
- При детальному аналізі можна спостерігати роки (2019-2021), коли виручка суттєво зросла або зменшилася. Це можуть бути важливі події або фактори, які вплинули на виручку.

**Пропозиції для можливого вдосконалення моделі прогнозування лінії тренду:**

Врахування інших факторів, таких як економічні індикатори або конкурентна ситуація, для більш точного прогноз.


Використання більш складних моделей, таких як експоненційне згладжування, для кращого аналізу сезонності і iн.

Врахування зовнішніх факторів: Якщо є доступ до додаткових даних, які можуть впливати на виручку (наприклад, економічні індикатори, рекламні витрати тощо), вони можуть бути включені в модель як екзогенні змінні для більш точного прогнозу.н.

## Прогноз За Допомогою Лінійної Функції Та Функції Регресії

1. Створення моделі регресії та прогнозування виручки на наступні 3 роки

```python
# Перетворення років у масив цілих чисел
years_numeric = np.array(years).reshape(-1, 1)
revenue_numeric = np.array(revenue)

# Створення моделі лінійної регресії та навчання на історичних даних
model = LinearRegression()
model.fit(years_numeric, revenue_numeric)

# Прогноз наступних 3 років
future_years = np.array([2022, 2023, 2024]).reshape(-1, 1)
revenue_forecast = model.predict(future_years)

# Виведення результатів
print("Прогноз виручки на наступні 3 роки:")
for year, forecast in zip([2022, 2023, 2024], revenue_forecast):
    print(f"Рік {year}: {forecast:.2f} грн.")
```

2. Розрахунок тенденції та росту

```python
# Додавання константи для моделі
X = sm.add_constant(years)

# Побудова моделі регресії
model = sm.OLS(revenue, X).fit()

# Отримання параметрів моделі
intercept, slope = model.params

# Виведення результатів
print("Результати аналізу регресії:")
print(f"Тенденція (slope): {slope:.2f}")
print(f"Ріст (за одиницю року): {slope:.2f} грн.")
```

2. Ріст.  Розрахован  значення рост  також є позитивним і вказує на те, що виручка зростає приблизно на одну і ту ж величину щороку.
